```python
# Superstore Sales Analysis
## End-to-End Data Cleaning and EDA using Python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("train.csv")
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df['Order Date'] = pd.to_datetime(df['Order Date'], dayfirst=True)
df['Ship Date'] = pd.to_datetime(df['Ship Date'], dayfirst=True)

df.info()
```

```python
df['Order Year'] = df['Order Date'].dt.year
df['Order Month'] = df['Order Date'].dt.month
df['Order Month Name'] = df['Order Date'].dt.month_name()
df.head()
```

```python
df['Shipping Duration'] = (df['Ship Date'] - df['Order Date']).dt.days

df[['Order Date', 'Ship Date', 'Shipping Duration']].head()
```

```python
df['Shipping Duration'].describe()
```

```python
df = df.dropna()
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
df.shape
```

```python
df['Sales'].sum()
```

```python
df['Sales'].mean()
```

```python
df['Sales'].median()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df['Sales'], bins=50)
plt.title("Sales Distribution")
plt.show()
```

```python
df.groupby('Category')['Sales'].sum().sort_values(ascending=False)
```

```python
plt.figure(figsize=(8,5))
sns.barplot(x='Category', y='Sales', data=df, estimator=sum)
plt.title("Sales by Category")
plt.show()
```

```python
df.groupby('Sub-Category')['Sales'].sum().sort_values(ascending=False).head(10)
```

```python
df.groupby('Region')['Sales'].sum().sort_values(ascending=False)
```

```python
plt.figure(figsize=(8,5))
sns.barplot(x='Region', y='Sales', data=df, estimator=sum)
plt.title("Sales by Region")
plt.show()
```

```python
pd.pivot_table(df, 
               values='Sales', 
               index='Region', 
               columns='Category', 
               aggfunc='sum')
```

```python
df.groupby('Order Year')['Sales'].sum()
```

```python
plt.figure(figsize=(8,5))
sns.lineplot(x='Order Year', y='Sales', data=df, estimator=sum)
plt.title("Yearly Sales Trend")
plt.show()
```

```python
monthly_sales = df.groupby(['Order Year','Order Month'])['Sales'].sum().reset_index()

plt.figure(figsize=(10,6))
sns.lineplot(x='Order Month', y='Sales', hue='Order Year', data=monthly_sales)
plt.title("Monthly Sales Trend by Year")
plt.show()
```

```python
df.groupby('Product Name')['Sales'].sum().sort_values(ascending=False).head(10)
```

```python
df.groupby('Segment')['Sales'].sum().sort_values(ascending=False)
```
